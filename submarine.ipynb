# Importing the dependencies


```python
import numpy as np #used for making arrays
import pandas as pd #used for making dataframes
from sklearn.model_selection import train_test_split #splits data into test and split data
from sklearn.linear_model import LogisticRegression #this is a logistic regression model so we use this dependency
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import joblib
```

# data collection and data processing

```python
sonar_data = pd.read_csv('/content/sonar data.csv', header=None)
```

```python
sonar_data.head()
```

```python
#number of rows and columns
sonar_data.shape
```

# Feature extraction

```python
features = sonar_data.iloc[:, :-1]
```

```python
target = sonar_data.iloc[:, -1]
```

```python
print("Features(Amplitude values at different frequency points - Sonar signal reflection)")
print(features.head())
print("\nTarget(Rock or Mine)")
print(target.head())
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
# seperating data and labels
X = sonar_data.drop(columns = 60, axis = 1)
Y = sonar_data[60]
```

```python
print(X)
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data[60].value_counts().plot(kind='bar')
```

```python
print(Y)
```

# training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.1, stratify = Y, random_state = 1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

# Model training --> Logistic regression model


```python
model_1 = LogisticRegression()
```

# training the logistic regression model with training data

```python
model_1.fit(X_train, Y_train)
```

```python
model_2 = GaussianNB()
```

```python
model_2.fit(X_train, Y_train)
```

# Model evaluation

```python
X_train_prediction_1 = model_1.predict(X_train)
training_data_accuracy_1 = accuracy_score(X_train_prediction_1, Y_train)
X_test_prediction_1 = model_1.predict(X_test)
test_data_accuracy_1 = accuracy_score(X_test_prediction_1, Y_test)

print('Logistic Regression Model - Accuracy on training data:', training_data_accuracy_1)
print('Logistic Regression Model - Accuracy on test data:', test_data_accuracy_1)
```

```python
X_train_prediction_2 = model_2.predict(X_train)
training_data_accuracy_2 = accuracy_score(X_train_prediction_2, Y_train)
X_test_prediction_2 = model_2.predict(X_test)
test_data_accuracy_2 = accuracy_score(X_test_prediction_2, Y_test)

print('Gaussian Naive Bayes Model - Accuracy on training data:', training_data_accuracy_2)
print('Gaussian Naive Bayes Model - Accuracy on test data:', test_data_accuracy_2)
```
